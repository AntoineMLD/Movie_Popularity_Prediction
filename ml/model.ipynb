```python
# Liste des genres à supprimer
genres_a_supprimer = ['Court-métrage', 'Mystère', 'Historique', 'Comédie Musicale', 'Film noir', 'Famille', 'Science-fiction']

# Supprimer les lignes contenant les genres à supprimer
for genre in genres_a_supprimer:
    data = data[~data['genres'].str.contains(genre)]

# Réinitialiser l'index
data.reset_index(drop=True, inplace=True)
```

```python
# Liste des valeurs à supprimer
valeurs_a_supprimer = ['0', 'Mini-série télévisée', '2h 25min', 'Vidéo', 'Téléfilm', 'Série télévisée']

# Supprimer les lignes où la feature 'annee' prend les valeurs à supprimer
data = data[~data['annee'].isin(valeurs_a_supprimer)]
# display(data)

# X['annee'] = X['annee'].astype(int)
print(data['annee'].unique())
print(data['duree'].unique())
print(data['genres'].unique())
print(data['genres'].value_counts())
print(data['continent'].unique())
print(data['pegi'].unique())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from feature_engine.encoding import CountFrequencyEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np 

X = data.drop("entrees_fr", axis=1)
y = data.entrees_fr

cols_drop = ["studio", "semaine_fr", "titre", "langue", "pays", "director_x", "scenaristes", "casting_principal", "casting_complet", "actors", "director_y", "nationalite", "title", "titre_original"]
X = X.drop(cols_drop, axis=1)




# Définition des colonnes catégorielles et numériques
object_cols = []
categorical_cols = ['genres', 'continent', 'pegi']
numeric_cols = ['duree', 'annee']

inputing_process= ColumnTransformer([
    ("impute_most_frequent", SimpleImputer(strategy='most_frequent'), object_cols),
])

# Création du ColumnTransformer
preprocessing = ColumnTransformer([
    # ("impute_most_frequent", SimpleImputer(strategy='most_frequent'), object_cols),
    # ("frequency", CountFrequencyEncoder(encoding_method="frequency", missing_values="ignore"), object_cols),
    ("onehot", OneHotEncoder(), categorical_cols),
    ("impute", SimpleImputer(strategy='mean'), numeric_cols)
])



# Création du pipeline complet avec la régression linéaire
model = Pipeline([
    # ("inputing_process", inputing_process),
    ("preprocessing", preprocessing),
    ("regression", LinearRegression())
])

display(model)

# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

display(X_train)

# Entraînement du modèle
model.fit(X_train, y_train)

# Évaluation du modèle
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

print("Score d'entraînement:", train_score)
print("Score de test:", test_score)

pred_test = model.predict(X_test)

# Calcul de l'erreur quadratique moyenne (MSE)
mse = mean_squared_error(y_test, pred_test)

# Calcul de l'erreur absolue moyenne (MAE)
mae = mean_absolute_error(y_test, pred_test)

rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE) :", rmse)
print("Mean Absolute Error (MAE) :", mae)
```
