```python
import pandas as pd

# Définir le taux de conversion en dollars pour chaque devise
taux_de_conversion = {
    '£': 1.38,  # Livre sterling (taux de change actuel)
    '€': 1.10,  # Euro (taux de change actuel)
    '₹': 0.014,  # Roupie indienne (taux de change actuel)
    'DEM': 0.62,  # Deutsche Mark (taux de change historique)
    '¥': 0.0093,  # Yen japonais (taux de change actuel)
    'FRF': 0.18,  # Franc français (taux de change historique)
    '₩': 0.00089,  # Won sud-coréen (taux de change actuel)
    'ATS': 0.090,  # Schilling autrichien (taux de change historique)
    'RUR': 0.013,  # Rouble russe (taux de change actuel)
    'DKK': 0.16,  # Couronne danoise (taux de change actuel)
    'CN¥': 0.15,  # Yuan chinois (taux de change actuel)
    'SEK': 0.12,  # Couronne suédoise (taux de change actuel)
    'ITL': 0.00061  # Lire italienne (taux de change historique)
}

# Fonction pour convertir le budget en dollars
def convertir_en_dollars(row):
    symbole = row['budget'].split()[0]  # Extraire le symbole (premier mot de la chaîne)
    montant = float(row['budget'].split()[1].replace(',', ''))  # Extraire le montant et supprimer les virgules
    if symbole == '$':
        return montant  # Déjà en dollars
    elif symbole in taux_de_conversion:
        taux = taux_de_conversion[symbole]
        return montant * taux
    else:
        return None  # Impossible de convertir

# Appliquer la conversion
data_imdb['budget_en_dollars'] = data_imdb.apply(convertir_en_dollars, axis=1)

# Afficher les premières lignes pour vérification
print(data_imdb[['budget', 'budget_en_dollars']].head())
```
