```python
import pandas as pd


df = pd.read_csv("moviesdata.csv")

# Dropping 'url' and 'salles_premiere_semaine' columns
movies_data_cleaned = df.drop(['url'], axis=1)

# Keeping rows where both 'acteurs' and 'budget' are not null to avoid incomplete data
movies_data_cleaned = movies_data_cleaned.dropna(subset=['acteurs', 'budget'])

#Dropping duplicate rows based on all columns to ensure unique movie entries
movies_data_cleaned = movies_data_cleaned.drop_duplicates()

movies_data_cleaned = movies_data_cleaned.groupby('titre').agg({
    'acteurs': 'first',  # Prendre la première valeur non nulle
    'budget': 'first',   # Prendre la première valeur non nulle
    'compositeur': 'first',  # Prendre la première valeur non nulle
    'date': 'first',     # Prendre la première valeur non nulle
    'date_US': 'first',  # Prendre la première valeur non nulle
    'duree': 'first',    # Prendre la première valeur non nulle
    'entrees_premiere_semaine': 'first',  # Prendre la première valeur non nulle
    'franchise': 'first',  # Prendre la première valeur non nulle
    'genre': 'first',     # Prendre la première valeur non nulle
    'pays': 'first',      # Prendre la première valeur non nulle
    'pegi_FR': 'first',   # Prendre la première valeur non nulle
    'pegi_US': 'first',   # Prendre la première valeur non nulle
    'producteur': 'first',  # Prendre la première valeur non nulle
    'realisateur': 'first',  # Prendre la première valeur non nulle
    'recette_premier_wk_US': 'first',  # Prendre la première valeur non nulle
    'remake': 'first',    # Prendre la première valeur non nulle
    'salles_premiere_semaine': 'first',  # Prendre la première valeur non nulle
    'studio': 'first'     # Prendre la première valeur non nulle
}).reset_index()

movies_data_cleaned.head()

# movies_data_cleaned.to_csv("movies_csv_uncleaned.csv")
```

```python
movies_data_cleaned.shape 
```

### Titre

```python
for index, value in movies_data_cleaned['titre'].items():
    if isinstance(value,str):
        value = value.replace('\r\n','')
        value = value.strip()
        movies_data_cleaned.at[index, 'titre'] = value
```

### Acteurs

```python
movies_data_cleaned['acteurs'] = movies_data_cleaned['acteurs'].str.strip().replace(r"\r\n", "", regex=True).replace(r"\s+", " ", regex=True) 

movies_data_cleaned.head()
```

### Studio

```python
for index, value in movies_data_cleaned['studio'].items():
    if isinstance(value,str):
        value = value.replace('\r\n','')
        value = value.strip()
        movies_data_cleaned.at[index, 'studio'] = value
```

### Réalisateur

```python
for index, value in movies_data_cleaned['realisateur'].items():
    if isinstance(value,str):
        value = value.replace('\r\n','')
        value = value.replace('-','')
        value = value.strip()
        movies_data_cleaned.at[index, 'realisateur'] = value

movies_data_cleaned.head() 
```

### Date

```python
movies_data_cleaned['date'] = movies_data_cleaned['date'].str.strip()
movies_data_cleaned['date'] = pd.to_datetime(movies_data_cleaned['date'], format='%d/%m/%Y')

movies_data_cleaned.head() 
```

### Date US

```python
for index, value in movies_data_cleaned['date_US'].items():
    if isinstance(value,str):
        value = value.replace('\r\n','')
        value = value.strip()
        movies_data_cleaned.at[index, 'date_US'] = value


movies_data_cleaned.head() 
```

### Budget

```python
movies_data_cleaned['budget'] = movies_data_cleaned['budget'].str.replace('[\$,€]', '', regex=True).str.replace(' ', '').str.replace('?', '')
movies_data_cleaned['budget'] = pd.to_numeric(movies_data_cleaned['budget'], errors='coerce').fillna(0).astype(int)

movies_data_cleaned.head()
```

### Nombre d'entrées

```python
movies_data_cleaned['entrees_premiere_semaine'] = movies_data_cleaned['entrees_premiere_semaine'].str.replace(' ', '').astype(int) 
```
